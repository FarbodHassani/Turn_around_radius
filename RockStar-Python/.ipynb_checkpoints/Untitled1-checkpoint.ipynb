```python
import numpy as np
import matplotlib.pylab as plt
import h5py as h5
%matplotlib inline
add_sing='../gevolution-dev-master-single/output-64grid-single/'
add_double='../gevolution-dev-master-double/output-64-double/'
red_shift=[48.5,29.8,10,2.9,1,0.45,0]
#for snap in range(7):
snap=6
power_s=np.loadtxt(add_sing+'lcdm_pk00'+str(snap)+'_phi.dat')
power_d=np.loadtxt(add_double+'lcdm_pk00'+str(snap)+'_phi.dat')
#if (power_d[:l,1]-power_s[:l,1]>0)


#axarr = plt.subplots(2, sharex=True)
#axarr[0].plot(power_d[:l,0], power_d[:l,0])
#axarr[0].set_title('Sharing X axis')
#axarr[1].scatter(power_d[:l,0], power_d[:l,0])


plt.subplot(211)
l=min(power_d.shape[0],power_s.shape[0])
plt.semilogy(power_d[:l,0],(power_d[:l,1]-power_s[:l,1]),'ro',label=r'$\frac{P_d-P_s}{P_d}$')
plt.semilogy(power_d[:l,0],-(power_d[:l,1]-power_s[:l,1]),'bo',label=r'$\frac{P_s-P_d}{P_d}$')
plt.xlim(0.19,0.25)
#plt.xlabel(r'$k[h/Mpc]$',fontsize=14)
#plt.ylim(1.e-15,1.e-13)
plt.ylabel(r'$\mathcal{P}_{\Phi}[Mpc^3/h^3]$',fontsize=10)
#plt.legend(loc='best',fontsize=10)
plt.legend(loc='upper center',fontsize=8, bbox_to_anchor=(0.95, 1.06),
          ncol=1, fancybox=True, shadow=True)
#plt.semilogy(power_d[:l,0],power_d[:l,1],'-bo',label='double')
#plt.axis('off')
plt.tick_params(
    axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom='off',      # ticks along the bottom edge are off
    top='off',         # ticks along the top edge are off
    labelbottom='off') 

plt.subplot(212)
#plt.semilogy(power_d[:l,0],power_s[:l,1],'-ro',label='Single')
plt.plot(power_d[:l,0],power_d[:l,1],'r*',markersize=6,label=r'$P_{d}$')
plt.plot(power_d[:l,0],power_s[:l,1],'bo',markersize=3,label=r'$P_{s}$')
plt.xlim(0.19,0.25)
#if (power_d[:l,1]-power_s[:l,1]<0)
#plt.semilogy(power_d[:l,0],(power_d[:l,1]-power_s[:l,1])/power_d[:l,1],label=r'$z='+str(red_shift[snap])+'$')

plt.xlabel(r'$k[h/Mpc]$',fontsize=10)
plt.xlim(0.19,0.25)
plt.ylim(1.e-13*3,1.e-12*2)
#plt.ylim(1.e-15,1.e-13)
plt.ylabel(r'$\mathcal{P}_{\Phi}[Mpc^3/h^3]$',fontsize=10)
plt.legend(loc='upper center',fontsize=8, bbox_to_anchor=(0.95, 1.06),
          ncol=1, fancybox=True, shadow=True)
plt.subplots_adjust(left=0.18, bottom=0.15, right=None, top=None)


#plt.figure('Power_Phi1.jpg',figsize=(30, 24))
plt.savefig("Power_Phi1.jpg",figsize=(40, 24), dpi = 400)

plt.show()
```
