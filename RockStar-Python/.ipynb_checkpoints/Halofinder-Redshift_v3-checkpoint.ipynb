```python
import numpy as np
import matplotlib.pylab as pltimport 
import h5py
#from mayavi import mlab

ff= h5py.File('lcdm_snap-01_cdm.h5',"r")


np.array(ff['data'])['positionX']

import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
s = np.array(list(ff['data']))
s['velocityX'].shape

import h5py
import matplotlib.pylab as plt


from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d import axes3d
fig = plt.figure()
n=1
ax = fig.gca(projection='3d')
xs = s['positionX'][:n]
ys = s['positionY'][:n]
zs = s['positionZ'][:n]

ax.scatter(xs, ys, zs, s=1, c='r', marker='o')

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```

```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np


fig = plt.figure()
ax = fig.gca(projection='3d')

xs = s['positionX'][:n]
ys = s['positionY'][:n]
zs = s['positionZ'][:n]
ux = s['velocityX'][:n]
uy = s['velocityY'][:n]
uz = s['velocityZ'][:n]

ax.quiver(xs, ys, zs, ux, uy, uz ,color='r',length=0.3)
ax.set_xlim(0,1)
ax.set_ylim(0,1)
ax.set_zlim(0,1)
plt.show()
```
